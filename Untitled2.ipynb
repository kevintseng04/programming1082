台灣自然增加率、出生率、死亡率-折線圖 (民國70年至108年)

```python
import pandas as pd
import matplotlib.pyplot as plt
tdata = pd.read_excel("人口變化.xlsx")

x1=tdata['年']
y1=tdata['自然增加率(0/00)']
plt.plot(x1,y1,label="Natural Increase Rate")


x2=tdata['年']
y2=tdata['粗出生率(0/00)']
plt.plot(x2,y2,label="Birth Rate")

x3=tdata['年']
y3=tdata['粗死亡率(0/00)']
plt.plot(x3,y3,label="Death Rate")

plt.title("RATE(Taiwan)")
plt.xlabel("Year")
plt.ylabel("Rate(‰)")

plt.legend()
plt.show()
```

台灣嬰兒出生數結構按生母胎次分-折線圖 (民國97至107年)

台灣不想生育因素(經濟負擔太重) -圓餅圖

```python
tdata = pd.read_excel("女性教育.xlsx")
counts = tdata["經濟負擔太重"]
colors = ['gold', 'yellowgreen', 'lightcoral','lightblue']
labels = tdata["教育程度"]

patches, texts = plt.pie(counts, colors = colors, startangle = 90,  autopct = "%1.1f%%", labels=labels)


plt.legend(patches, loc = "best")

plt.axis('equal')
plt.tight_layout()
plt.show()

```
